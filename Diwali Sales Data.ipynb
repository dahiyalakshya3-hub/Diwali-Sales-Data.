```python
import pandas as pd
import numpy as np

df = pd.read_csv("Diwali Sales Data.csv")

# total sales
total_sales = np.sum(df['Amount'])

print(total_sales)
```

```python
df.groupby('Gender')['Amount'].sum()
np.mean(df['Amount'])
```

```python
import matplotlib.pyplot as plt
df.groupby('State')['Amount'].sum().plot(kind='bar')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Gender Analysis
```

```python
sales_gender = df.groupby('Gender')['Amount'].sum()
print(sales_gender)
```

```python
sales_gender.plot(kind='bar', color=['blue','pink'])
plt.title("Gender vs Sales")
plt.show()
```

```python
Age Group Analysis
```

```python
df.groupby('Age Group')['Amount'].sum().plot(kind='bar')
plt.title("Age Group vs Sales")
plt.show()
```

```python
State-wise Analysis
```

```python
top_states = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)

top_states.plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.show()
```

```python
Product Category Analysis
```

```python
df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Product Categories")
plt.show()
```

```python
NumPy Insights
```

```python
print("Average Purchase:", np.mean(df['Amount']))
print("Total Sales:", np.sum(df['Amount']))
```

```python
Final Insights (Conclusion)
```
